```python
import os

train_img_dir = "data/images/train"
train_images = set(os.listdir(train_img_dir))
```

```python
from pathlib import Path

def get_image_filename(url):
    if isinstance(url, str):
        return Path(url).name
    return None

train_df["image_filename"] = train_df["image_link"].apply(get_image_filename)
```

```python
train_df["has_image"] = train_df["image_filename"].isin(train_images).astype(int)
```

```python
train_df["has_image"].value_counts(normalize=True)
```

Using a pretrained CNN that is EfficientNetB0, we will extract features from the images and use them to predict the price of the product.

```python
#using image_features.py to extract features from images
from src.image_features import extract_embeddings

train_filenames = train_df.loc[
    train_df["has_image"] == 1, "image_filename"
].tolist()

train_embeddings = extract_embeddings(
    image_dir="data/images/train",
    filenames=train_filenames,
    device="cpu"   # GPU if available
)
```

```python
import pickle

with open("data/train_image_embeddings.pkl", "wb") as f:
    pickle.dump(train_embeddings, f)
```
